Create a class to represent the **Investor**. Basic information to be provided: full information about the investor
Main implemented behavior: access to the domain name and adding or change domain name

```python
import json
```

```python
class Investor:
  """
  The investor class
  """
  def __init__(self, name: str, practice_area_list: str, description: str,
               tracxn_investment_score: int, domain: str):
    """
    Constructor
    """
    self.name = name
    self.practice_area_list = practice_area_list
    self.tracxn_investment_score = tracxn_investment_score
    self.domain = domain
    self.description = description

  def __str__(self):
    """
    Str appearance
    """
    return f'{self.name}, Description: {self.description}, Domain: {self.domain}'

  @classmethod
  def create_investor(cls, data: str, index):
    """
    Creating class instance throught dataset
    """
    with open(data, 'r') as file:
      data1 = json.loads(file.read().replace("'", '"').replace('True', 'true')\
       .replace('False', 'false'))['result'][index]

    return cls(data1['name'], data1['practiceAreaList'], data1['description']['short'],
               int(data1['tracxnInvestmentScore']), data1['domain'])
```

**Exercise: Scrape Data From a Website**

Write a program that grabs the full HTML from the following URL:

```python
url = "http://olympus.realpython.org/profiles/dionysus"
```

Then use `.find()` to display the text following Name: and Favorite Color: (not including any leading spaces or trailing HTML tags that might appear on the same line).

####You can expand the block below to see a solution.

```python
from urllib.request import urlopen

page = urlopen(url)
html = page.read().decode('utf-8')

for quality in ["Name: ", "Favorite Color:"]:
  start_pos = html.find(quality) + len(quality)
  end_pos = html[start_pos:].find("<")

  print(html[start_pos:end_pos + start_pos].strip(" \r\n\t"))
```

**Exercise: Parse HTML With Beautiful Soup**

Write a program that grabs the full HTML from the page at the URL `http://olympus.realpython.org/profiles`.

Using Beautiful Soup, print out a list of all the links on the page by looking for HTML tags with the name `a` and retrieving the value taken on by the `href` attribute of each tag.

The final output should look like this:

```python
http://olympus.realpython.org/profiles/aphrodite
http://olympus.realpython.org/profiles/poseidon
http://olympus.realpython.org/profiles/dionysus
```

Make sure that you only have one slash `(/)` between the base URL and the relative URL.

```python
from urllib.request import urlopen
from bs4 import BeautifulSoup

url = "http://olympus.realpython.org"

page = urlopen(url + "/profiles")
html = page.read().decode('utf-8')

soup = BeautifulSoup(html, 'html.parser')
for l in soup.find_all('a'):
  print(url + l['href'])
```

**Exercise: Submit a Form With Mechanical Soup**

Use MechanicalSoup to provide the correct username `(zeus)` and password `(ThunderDude)` to the login form located at the URL `http://olympus.realpython.org/login`.

Once the form is submitted, display the title of the current page to determine that you’ve been redirected to the `/profiles` page.

Your program should print the text `<title>All Profiles</title>`.


```python
import mechanicalsoup

url = "http://olympus.realpython.org/login"

browser = mechanicalsoup.Browser()
page = browser.get(url)
html = page.soup

form = html.form

form('input')[0]['value'] = 'zeus'
form('input')[1]['value'] = 'ThunderDude'

submit = browser.submit(form, page.url)
print(submit.soup.title)
```
